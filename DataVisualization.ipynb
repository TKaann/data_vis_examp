### Date kolonundan yeni kolonlar ve veriler olusturma. (Date kolonunu objectten datetime formatina cektigimiz icin olusturabiliyoruz)

```python
df['day'] = (df['Date']).dt.day    #date kolonundan yeni kolonlar olusturuyoruz. df['day'] ile kolonumuza isim verdik
#dt.day ile de datetime formatindan day ozelligini tasiyan degeri cektik.
df['month'] = (df['Date']).dt.month    #aynisini month
df['year'] = (df['Date']).dt.year   #ve year icin yaptik
#datamizin sonunda day month ve year olarak 3 yeni kolon actik
#bunu yapabilmemizin sebebi Date kolonumuzu DateTime formatina cevirebilmemizden kaynakli.
```

```python
df.sample()
```

```python
df['month_name'] = (df['Date']).dt.month_name()   #Date kolonundan month_name adinda yeni bir kolon olusturuyorum.
#.dt.month_name() ile ayin ismini yeni kolonumuza yazdiriyoruz
df['weekday'] = (df['Date']).dt.day_name()  #ayni sekilde bunu gunler icin de yapiyoruz. 
```

```python
df.sample()   #verisetimizden ornek satirlar cekiyoruz
```

```python
df['Time'] = pd.to_datetime(df['Time'])   #Time kolonumuzu da datetime formatina ceviriyoruz. onceden objeydi
#pd.to_datetime  formatini kullanip tur donusumlerimizi gerceklestiriyoruz.
#ayni degiskenler gibi mesela int = sayi tutuyor, string = kelime tutuyor, char = karakter tutuyor,
#gibi degiskenlerimizin spesifik ozellikleri var.
#datetime formatinin da dilin o sayilari anlamlandirip ay yil gun saat diye anlamlandirmabilmesine yariyor gibi dusunebilirsiniz. 
#o sayede ustunde oynamalar yapabiliyoruz
```

```python
import matplotlib.pyplot as plt   #matplotlib kutuphanesini import ettik
import warnings     #aldigimiz uyarilari gostermemesi icin warnings'i import ediyoruz
warnings.filterwarnings('ignore')   #warningleri filtrele ve ignore et
```

```python
#df.groupby('month_name'): Bu ifade, 'month_name' stununa gre verileri gruplar. Yani, ayn ay ismine sahip tm satrlar bir araya getirir.
#.size(): Her grup iin boyutu (yani, her ayn ka tane sat olduunu) hesaplar. Bu, her ayn sat saysn elde etmek iin kullanlr.
#.reset_index(name='count'): Bu adm, gruplama ilemi sonucunda elde edilen veriyi bir DataFrame'e dntrr.
#'name' parametresi, yeni stunun adn belirler. Bu durumda, sat saysnn yer ald yeni stunun ad 'count' olarak belirlenir. 
#Bu ilem, gruplanm verileri yeniden dzenler ve daha kolay kullanlabilir hale getirir.
#Sonu olarak, bu kod, 'month_name' stunu tarafndan gruplanm ve her ayn sat saysn ieren bir DataFrame'i oluturur.
#'count' adl yeni stun, her ayn sat saysn ierir ve bu veri ubuk grafik oluturmak gibi ilemlerde kullanlabilir.
```
